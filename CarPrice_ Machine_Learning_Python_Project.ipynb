```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardscaler=StandardScaler()
standardscaler.fit_transform(x)
```

```python
x=pd.DataFrame(standardscaler.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100 , lr.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,lr.predict(x_test)*100 )
```

```python
lr1=Lasso(alpha=0.05)
lr1.fit(x_train,y_train)
lr1.score(x_train,y_train)*100 , lr1.score(x_test,y_test)*100
```

```python
lr2=Ridge(alpha=0.05)
lr2.fit(x_train,y_train)
lr2.score(x_train,y_train)*100 , lr2.score(x_test,y_test)*100
```

```python
lr3=ElasticNet(alpha=0.05)
lr3.fit(x_train,y_train)
lr3.score(x_train,y_train)*100 , lr3.score(x_test,y_test)*100
```

```python
Dt=DecisionTreeRegressor()
Dt.fit(x_train,y_train)
Dt.score(x_train,y_train)*100 , Dt.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,Dt.predict(x_test)*100)
```

```python
Rf=RandomForestRegressor()
Rf.fit(x_train,y_train)
Rf.score(x_train,y_train)*100 , Rf.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,Rf.predict(x_test)*100 )
```

```python
Sv=SVR()
Sv.fit(x_train,y_train)
Sv.score(x_train,y_train)*100 , Sv.score(x_test,y_test)*100
```

```python
Kn=KNeighborsRegressor(n_neighbors=2)
Kn.fit(x_train,y_train)
Kn.score(x_train,y_train)*100 , Kn.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,Kn.predict(x_test)*100) 
```
