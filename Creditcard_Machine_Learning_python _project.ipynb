```python
from imblearn.over_sampling import SMOTE
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
smote=SMOTE(random_state=42)

x_train,y_train=smote.fit_resample(x_train,y_train)
```

```python
print("after smoting the trainning  dataset")
pd.Series(y_train).value_counts()
```

```python
x_test,y_test= smote.fit_resample(x_test,y_test)
print("after smoting the testing dataset")
pd.Series(y_test).value_counts()
```

```python
print("trainning data set size is",len(x_train))
print("class balance is",sum(y_train))
print("testing data set size is",len(x_test))
print("class balance is",sum(y_test))
```
